# Pie chart

```python
import matplotlib.pyplot as plt

# Market share data
manufacturers = ['Samsung', 'Apple', 'Huawei', 'Xiaomi', 'Others']
market_share = [30, 25, 18, 12, 15]

# Colors for each slice
colors = ['lightblue', 'lightcoral', 'lightgreen', 'lightpink', 'lightgrey']
# Create a pie chart

plt.pie(market_share, 
        labels=manufacturers, 
        colors=colors, 
        autopct='%1.1f%%',
        wedgeprops = {"edgecolor" : "black", 
                      'linewidth': 1.8, 
                      'antialiased': True})

# Title
plt.title('Smartphone Market Share')
# Show the pie chart
plt.show()
```

# Scatter Plot

```python
import matplotlib.pyplot as plt

# Student data (study hours and exam scores)
study_hours = [2, 3, 1, 4, 3, 5, 2, 6, 5, 7]
exam_scores = [65, 75, 60, 80, 70, 85, 70, 90, 88, 92]

# Create a scatter plot
scat = plt.scatter(study_hours, exam_scores, c='blue', marker='D', label='Student Data')

# Labeling and Title
for i, sc in enumerate(exam_scores):
    plt.text(study_hours[i] + 0.1, exam_scores[i], str(sc), fontsize=9)
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')
plt.title('Study Hours vs. Exam Scores')

# Show the scatter plot
plt.legend()
plt.grid(True)
plt.show()
```

# Subplots

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample student data
students = ['Jhon', 'Smith', 'Marry', 'Rose', 'Devid']
math_scores = [85, 92, 78, 88, 90]
science_scores = [76, 88, 92, 80, 78]
english_scores = [82, 89, 91, 84, 77]
history_scores = [75, 85, 80, 82, 88]

# Create a subplot with four subplots (2 rows, 2 columns)
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Plot Math Scores (Top-left)
bar1 = axs[0, 0].bar(students, math_scores, color='b')
axs[0, 0].set_title('Math Scores')
axs[0, 0].set_xlabel('Students')
axs[0, 0].set_ylabel('Score')
axs[0,0].bar_label(bar1, labels=math_scores)

# Plot Science Scores (Top-right)
bar2 = axs[0, 1].bar(students, science_scores, color='g')
axs[0, 1].set_title('Science Scores')
axs[0, 1].set_xlabel('Students')
axs[0, 1].set_ylabel('Score')
axs[0,1].bar_label(bar2, labels=science_scores)

# Plot English Scores (Bottom-left)
bar3 = axs[1, 0].bar(students, english_scores, color='r')
axs[1, 0].set_title('English Scores')
axs[1, 0].set_xlabel('Students')
axs[1, 0].set_ylabel('Score')
axs[1,0].bar_label(bar3, labels=english_scores)


# Plot History Scores (Bottom-right)
bar4 = axs[1, 1].bar(students, history_scores, color='orange')
axs[1, 1].set_title('History Scores')
axs[1, 1].set_xlabel('Students')
axs[1, 1].set_ylabel('Score')
axs[1,1].bar_label(bar4, labels=history_scores)


# Adjust spacing between subplots
plt.tight_layout()

# Show the subplot
plt.show()
```

# Subplots 2

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample student data
students = ['Jhon', 'Smith', 'Marry', 'Rose', 'Devid']
math_scores = [85, 92, 78, 88, 90]
science_scores = [76, 88, 92, 80, 78]
english_scores = [82, 89, 91, 84, 77]
history_scores = [75, 85, 80, 82, 88]

# Create a subplot with four subplots (2 rows, 2 columns)
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Plot Math Scores (Top-left)
axs[0, 0].plot(students, math_scores, color='b', marker='o', linestyle='-')
axs[0, 0].set_title('Math Scores')
axs[0, 0].set_xlabel('Students')
axs[0, 0].set_ylabel('Score')
axs[0,0].grid()
for i, score in enumerate(math_scores):
    axs[0, 0].text(students[i], math_scores[i] + 0.5, str(score), ha='center')

# Plot Science Scores (Top-right)
axs[0, 1].plot(students, science_scores, color='g', marker='o', linestyle='-')
axs[0, 1].set_title('Science Scores')
axs[0, 1].set_xlabel('Students')
axs[0, 1].set_ylabel('Score')
axs[0,1].grid()

for i, score in enumerate(science_scores):
    axs[0, 1].text(students[i], science_scores[i] + 0.5, str(score), ha='center')

# Plot English Scores (Bottom-left)
axs[1, 0].plot(students, english_scores, color='r', marker='o', linestyle='-')
axs[1, 0].set_title('English Scores')
axs[1, 0].set_xlabel('Students')
axs[1, 0].set_ylabel('Score')
axs[1,0].grid()

for i, score in enumerate(english_scores):
    axs[1, 0].text(students[i], english_scores[i] + 0.5, str(score), ha='center')

# Plot History Scores (Bottom-right)
axs[1, 1].plot(students, history_scores, color='orange', marker='o', linestyle='-')
axs[1, 1].set_title('History Scores')
axs[1, 1].set_xlabel('Students')
axs[1, 1].set_ylabel('Score')
axs[1,1].grid()

for i, score in enumerate(history_scores):
    axs[1, 1].text(students[i], history_scores[i] + 0.5, str(score), ha='center')

# Adjust spacing between subplots
plt.tight_layout()

# Show the subplot
plt.show()
```
