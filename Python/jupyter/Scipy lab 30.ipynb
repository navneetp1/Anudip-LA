```python
import numpy as np
import scipy.io.wavfile as wav
import matplotlib.pyplot as plt

sampleRate, audioData = wav.read("audio_file.wav")

startSample = 1000
segmentLength = 1024
audioSegment = audioData[startSample:startSample+segmentLength]

fftRes = np.fft.fft(audioSegment)
frequencies = np.fft.fftfreq(len(fftRes),1/sampleRate)

magnitude = np.abs(fftRes)

plt.plot(frequencies, magnitude)

plt.xlabel("Frequency(Hz)")
plt.ylabel("Magnitude")

plt.title("'Frequency Spectrum of the Audio Segment")

plt.grid()
plt.show()
```

## Interpolation is a technique used to estimate data points between existing data points. constructing new aata points within the range


#### given the data of 9 days temperature with 5th day missing, using interpolation, we can make a rough prediction of day 5 temperature

```python
import numpy as np
from scipy.interpolate import interp1d

daysData = [1,2,3,4,6,7,8,9,10]
temperaturesData = [30,32,31,25,27,38,39,33,34]

days = np.array(daysData)
temperatures = np.array(temperaturesData)


desiredDay = 5
interpolationFunc = interp1d(days, temperatures, kind="linear")
estimatedTemp = interpolationFunc(desiredDay)

plt.plot(days, temperatures,"o", label="Measured Temperatures")
plt.plot(desiredDay, estimatedTemp, "s", label="Estimated Temperature")

plt.xlabel("Days")
plt.ylabel("Temperatures(°C)")

for i, sc in enumerate(temperatures):
    plt.text(days[i] + 0.1, temperatures[i], str(sc), fontsize=9)

plt.text(desiredDay+0.1,estimatedTemp,f"{estimatedTemp:}",fontsize=9)

plt.title("Temperature Interpolation")

plt.legend()
plt.grid()
plt.show()

print(f"Estimated Temperature of Day {desiredDay} is {estimatedTemp}°C")
```

```python
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
X = np.arange(11)
numbers = [2.0, 1.9, 1.7, 1.5, 0.5, 0.0, 0.8, 2.0, 0.9, 0.4, 2.0]
Y = np.array(numbers)
plt.plot(X,Y,"o:") # dotted lines connection
plt.show()

predict = interp1d(X,Y,kind="quadratic")
# to find x2 - input   with known values y2 - output

X2 = np.linspace(0,10,100) # 100 values evenly spaced b/w 0 and 10

Y2 = np.array([predict(res) for res in X2])

plt.plot(X2,Y2, "ro:")
plt.show()


```
